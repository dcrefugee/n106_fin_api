```python
### Pull my current tickers in from sorted sheet
# read 2nd sheet of an excel file
bp_tickers = pd.read_excel(r'O:\_Tiller on One Drive\Equity Dividend Tracker\tickers_for_yfinance.xlsx', sheet_name = "current_unique_bp_tickers")
#bp_tickers.head()

### Make my tickers df sybols column into a list
# Converting DataFrame to a list containing
# all the rows of column 'Name'
tickers = bp_tickers['symbol'].tolist()
bp_dow_list = tickers
 
# Printing the converted list.
#print(bp_dow_list)

### Retrieve basic ticker info and the try/pass skips exceptions
dow_stats_basic = {}
for ticker in tickers:
    try:
        temp = si.get_quote_data(ticker)
        dow_stats_basic[ticker] = temp
    except:
        pass
#dow_stats_basic

### Convert to a Pandas DataFrame

dow_stats_pd = pd.DataFrame.from_dict(dow_stats_basic)
dow_stats_pd2 = pd.DataFrame.transpose(dow_stats_pd)
dow_stats_pd2['date'] = pd.to_datetime("today").strftime("%m/%d/%Y")
#dow_stats_pd2

### To SQLite on local drive

# Creates and/or connects to SQLite DB Named lotto_results1.db
connection = db.connect(r'P:\Luke Barosso Python\db_house\wsb_df_test.db')

#Appends the above DF to the DB Created into a table named mega_millions1_df
dow_stats_pd2.to_sql(name="dow_stats_pd2_df", con=connection, if_exists='append', index=False)

## Commit connection
connection.commit()
```
